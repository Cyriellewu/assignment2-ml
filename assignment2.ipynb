```python
# =========================================================
# Assignment 2: From Trees to Neural Networks
# Dataset: Home Credit Default Risk (application_train.csv)
# Improved full version for Kaggle / GitHub submission
# =========================================================

# =========================
# 0. Imports / Setup
# =========================
import os
import time
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import sparse

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    average_precision_score,
    classification_report,
    precision_recall_curve
)
from sklearn.neural_network import MLPClassifier

from xgboost import XGBClassifier

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# =========================
# 1. Paths
# =========================
RESULTS_DIR = "./results"
FIGURES_DIR = os.path.join(RESULTS_DIR, "figures")
os.makedirs(RESULTS_DIR, exist_ok=True)
os.makedirs(FIGURES_DIR, exist_ok=True)

# Kaggle / local path fallback
POSSIBLE_DATA_PATHS = [
    "/kaggle/input/home-credit-default-risk/application_train.csv",
    "/kaggle/input/competitions/home-credit-default-risk/application_train.csv",
    "./data/application_train.csv",
    "./application_train.csv",
]

DATA_PATH = None
for p in POSSIBLE_DATA_PATHS:
    if os.path.exists(p):
        DATA_PATH = p
        break

if DATA_PATH is None:
    raise FileNotFoundError(
        "application_train.csv not found. Place it in ./data/ or run in Kaggle."
    )

print(f"Using DATA_PATH: {DATA_PATH}")

# =========================
# 2. Helper Functions
# =========================
def make_sparse_ohe():
    try:
        return OneHotEncoder(handle_unknown="ignore", sparse_output=True)
    except TypeError:
        return OneHotEncoder(handle_unknown="ignore", sparse=True)

def make_dense_ohe():
    try:
        return OneHotEncoder(handle_unknown="ignore", sparse_output=False)
    except TypeError:
        return OneHotEncoder(handle_unknown="ignore", sparse=False)

def evaluate_model(name, y_true, y_pred, y_prob, threshold=None):
    result = {
        "model": name,
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "f1": f1_score(y_true, y_pred, zero_division=0),
        "auc_pr": average_precision_score(y_true, y_prob),
    }
    if threshold is not None:
        result["threshold"] = threshold
    return result

def results_to_df(results_list, sort_by="auc_pr", ascending=False):
    df = pd.DataFrame(results_list).sort_values(by=sort_by, ascending=ascending).reset_index(drop=True)
    display(df)
    return df

def get_best_threshold_by_f1(y_true, y_prob):
    precision, recall, thresholds = precision_recall_curve(y_true, y_prob)

    # thresholds length = len(precision)-1
    # compute F1 only for real thresholds
    f1_scores = []
    for i in range(len(thresholds)):
        p = precision[i]
        r = recall[i]
        f1 = 0.0 if (p + r) == 0 else 2 * p * r / (p + r)
        f1_scores.append(f1)

    if len(f1_scores) == 0:
        return 0.5

    best_idx = int(np.argmax(f1_scores))
    return float(thresholds[best_idx])

def apply_threshold(y_prob, threshold):
    return (y_prob >= threshold).astype(int)

def get_pos_weight(y):
    pos = np.sum(y == 1)
    neg = np.sum(y == 0)
    return neg / max(pos, 1)

def save_current_figure(filename):
    path = os.path.join(FIGURES_DIR, filename)
    plt.tight_layout()
    plt.savefig(path, dpi=150, bbox_inches="tight")
    print(f"Saved figure to: {path}")

def plot_top_feature_importance(feature_importance_df, top_n=20, title="Top Feature Importance", filename=None):
    top_df = feature_importance_df.head(top_n).sort_values("importance", ascending=True)

    plt.figure(figsize=(10, 8))
    plt.barh(top_df["feature"], top_df["importance"])
    plt.title(title)
    plt.xlabel("Importance")
    plt.ylabel("Feature")
    if filename:
        save_current_figure(filename)
    plt.show()

# =========================
# 3. Load Data
# =========================
df = pd.read_csv(DATA_PATH)
print("Raw data shape:", df.shape)
display(df.head())

# =========================
# 4. Basic Cleaning / Feature Engineering
# =========================
TARGET_COL = "TARGET"

drop_cols = ["SK_ID_CURR"]
drop_cols = [c for c in drop_cols if c in df.columns]

if "DAYS_EMPLOYED" in df.columns:
    df["DAYS_EMPLOYED_ANOM"] = (df["DAYS_EMPLOYED"] == 365243).astype(int)
    df.loc[df["DAYS_EMPLOYED"] == 365243, "DAYS_EMPLOYED"] = np.nan

if set(["AMT_CREDIT", "AMT_INCOME_TOTAL"]).issubset(df.columns):
    df["CREDIT_INCOME_RATIO"] = df["AMT_CREDIT"] / (df["AMT_INCOME_TOTAL"] + 1e-6)

if set(["AMT_ANNUITY", "AMT_INCOME_TOTAL"]).issubset(df.columns):
    df["ANNUITY_INCOME_RATIO"] = df["AMT_ANNUITY"] / (df["AMT_INCOME_TOTAL"] + 1e-6)

if set(["AMT_CREDIT", "AMT_GOODS_PRICE"]).issubset(df.columns):
    df["CREDIT_GOODS_RATIO"] = df["AMT_CREDIT"] / (df["AMT_GOODS_PRICE"] + 1e-6)

if "DAYS_BIRTH" in df.columns:
    df["AGE_YEARS"] = (-df["DAYS_BIRTH"]) / 365.0

X = df.drop(columns=[TARGET_COL] + drop_cols)
y = df[TARGET_COL].copy()

print("Feature matrix shape:", X.shape)
print("Target mean:", y.mean())

# =========================
# 5. Train / Validation / Test Split
# =========================
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X,
    y,
    test_size=0.15,
    stratify=y,
    random_state=RANDOM_STATE
)

val_ratio_adjusted = 0.15 / 0.85  # final ~70/15/15
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full,
    y_train_full,
    test_size=val_ratio_adjusted,
    stratify=y_train_full,
    random_state=RANDOM_STATE
)

print("Train shape:", X_train.shape)
print("Val shape:", X_val.shape)
print("Test shape:", X_test.shape)

# =========================
# 6. Column Types
# =========================
categorical_cols = X_train.select_dtypes(include=["object"]).columns.tolist()
numeric_cols = X_train.select_dtypes(exclude=["object"]).columns.tolist()

print("Numeric columns:", len(numeric_cols))
print("Categorical columns:", len(categorical_cols))

# =========================
# 7. Preprocessors
# =========================
# XGBoost: impute + one-hot, no scaling
numeric_transformer_xgb = Pipeline([
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_transformer_xgb = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", make_sparse_ohe())
])

preprocessor_xgb = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer_xgb, numeric_cols),
        ("cat", categorical_transformer_xgb, categorical_cols)
    ]
)

# MLP: impute + scale numeric + one-hot categorical
numeric_transformer_mlp = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer_mlp = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", make_dense_ohe())
])

preprocessor_mlp = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer_mlp, numeric_cols),
        ("cat", categorical_transformer_mlp, categorical_cols)
    ]
)

# =========================
# 8. Fit Preprocessors on Train Only
# =========================
print("\nFitting preprocessors on training set only...")

X_train_xgb = preprocessor_xgb.fit_transform(X_train)
X_val_xgb = preprocessor_xgb.transform(X_val)
X_test_xgb = preprocessor_xgb.transform(X_test)

X_train_mlp = preprocessor_mlp.fit_transform(X_train)
X_val_mlp = preprocessor_mlp.transform(X_val)
X_test_mlp = preprocessor_mlp.transform(X_test)

feature_names_xgb = preprocessor_xgb.get_feature_names_out()
feature_names_mlp = preprocessor_mlp.get_feature_names_out()

print("X_train_xgb shape:", X_train_xgb.shape)
print("X_train_mlp shape:", X_train_mlp.shape)

# For final retraining later
X_trainval = pd.concat([X_train, X_val], axis=0)
y_trainval = pd.concat([y_train, y_val], axis=0)

# =========================
# 9. Class Imbalance Info
# =========================
xgb_scale_pos_weight = get_pos_weight(y_train)
print("scale_pos_weight for XGBoost:", xgb_scale_pos_weight)

# =========================
# 10. XGBoost Experiment 1: learning_rate
# =========================
print("\n" + "=" * 60)
print("XGBoost Experiment 1: learning_rate")
print("=" * 60)

learning_rates = [0.01, 0.1, 0.3]
xgb_lr_results = []
xgb_lr_histories = {}
xgb_lr_best_iters = {}

base_n_estimators = 1000

for lr in learning_rates:
    print(f"\nTraining XGBoost with learning_rate={lr}")

    model = XGBClassifier(
        n_estimators=base_n_estimators,
        learning_rate=lr,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.0,
        reg_lambda=1.0,
        scale_pos_weight=xgb_scale_pos_weight,
        objective="binary:logistic",
        eval_metric="logloss",
        random_state=RANDOM_STATE,
        tree_method="hist",
        n_jobs=-1,
        early_stopping_rounds=50
    )

    start_time = time.time()
    model.fit(
        X_train_xgb,
        y_train,
        eval_set=[(X_train_xgb, y_train), (X_val_xgb, y_val)],
        verbose=False
    )
    train_time = time.time() - start_time

    y_val_prob = model.predict_proba(X_val_xgb)[:, 1]
    best_threshold = get_best_threshold_by_f1(y_val, y_val_prob)
    y_val_pred = apply_threshold(y_val_prob, best_threshold)

    result = evaluate_model(
        name=f"XGB_lr_{lr}",
        y_true=y_val,
        y_pred=y_val_pred,
        y_prob=y_val_prob,
        threshold=best_threshold
    )
    result["learning_rate"] = lr
    result["train_time_sec"] = train_time
    result["best_iteration"] = int(model.best_iteration) if model.best_iteration is not None else base_n_estimators

    xgb_lr_results.append(result)
    xgb_lr_histories[lr] = model.evals_result()
    xgb_lr_best_iters[lr] = result["best_iteration"]

xgb_lr_df = results_to_df(xgb_lr_results)

plt.figure(figsize=(8, 5))
plt.plot(xgb_lr_df["learning_rate"], xgb_lr_df["auc_pr"], marker="o")
plt.title("XGBoost: Validation AUC-PR vs Learning Rate")
plt.xlabel("learning_rate")
plt.ylabel("Validation AUC-PR")
plt.grid(True)
save_current_figure("xgb_learning_rate_aucpr.png")
plt.show()

for lr in learning_rates:
    history = xgb_lr_histories[lr]
    train_loss = history["validation_0"]["logloss"]
    val_loss = history["validation_1"]["logloss"]

    plt.figure(figsize=(8, 5))
    plt.plot(train_loss, label="Train Logloss")
    plt.plot(val_loss, label="Validation Logloss")
    plt.title(f"XGBoost Loss Curve (learning_rate={lr})")
    plt.xlabel("Boosting Round")
    plt.ylabel("Logloss")
    plt.legend()
    plt.grid(True)
    save_current_figure(f"xgb_loss_curve_lr_{str(lr).replace('.', '_')}.png")
    plt.show()

best_lr_row = xgb_lr_df.iloc[0]
best_lr = float(best_lr_row["learning_rate"])
print("Best learning_rate:", best_lr)

# =========================
# 11. XGBoost Experiment 2: n_estimators
# =========================
print("\n" + "=" * 60)
print("XGBoost Experiment 2: n_estimators")
print("=" * 60)

n_estimators_list = [100, 300, 1000]
xgb_n_estimators_results = []

for n_est in n_estimators_list:
    print(f"\nTraining XGBoost with n_estimators={n_est}")

    model = XGBClassifier(
        n_estimators=n_est,
        learning_rate=best_lr,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.0,
        reg_lambda=1.0,
        scale_pos_weight=xgb_scale_pos_weight,
        objective="binary:logistic",
        eval_metric="logloss",
        random_state=RANDOM_STATE,
        tree_method="hist",
        n_jobs=-1
    )

    start_time = time.time()
    model.fit(X_train_xgb, y_train, verbose=False)
    train_time = time.time() - start_time

    y_val_prob = model.predict_proba(X_val_xgb)[:, 1]
    best_threshold = get_best_threshold_by_f1(y_val, y_val_prob)
    y_val_pred = apply_threshold(y_val_prob, best_threshold)

    result = evaluate_model(
        name=f"XGB_nest_{n_est}",
        y_true=y_val,
        y_pred=y_val_pred,
        y_prob=y_val_prob,
        threshold=best_threshold
    )
    result["n_estimators"] = n_est
    result["train_time_sec"] = train_time
    xgb_n_estimators_results.append(result)

xgb_n_estimators_df = results_to_df(xgb_n_estimators_results)

plt.figure(figsize=(8, 5))
plt.plot(xgb_n_estimators_df["n_estimators"], xgb_n_estimators_df["auc_pr"], marker="o")
plt.title("XGBoost: Validation AUC-PR vs n_estimators")
plt.xlabel("n_estimators")
plt.ylabel("Validation AUC-PR")
plt.grid(True)
save_current_figure("xgb_n_estimators_aucpr.png")
plt.show()

best_n_estimators = int(xgb_n_estimators_df.iloc[0]["n_estimators"])
print("Best explicit n_estimators:", best_n_estimators)

# =========================
# 12. XGBoost Experiment 3: other hyperparameters
# =========================
print("\n" + "=" * 60)
print("XGBoost Experiment 3: max_depth / subsample / reg_alpha / reg_lambda")
print("=" * 60)

xgb_tuning_results = []

param_settings = [
    {"max_depth": 4, "subsample": 0.8, "reg_alpha": 0.0, "reg_lambda": 1.0},
    {"max_depth": 6, "subsample": 0.8, "reg_alpha": 0.0, "reg_lambda": 1.0},
    {"max_depth": 8, "subsample": 0.8, "reg_alpha": 0.0, "reg_lambda": 1.0},
    {"max_depth": 6, "subsample": 0.6, "reg_alpha": 0.0, "reg_lambda": 1.0},
    {"max_depth": 6, "subsample": 1.0, "reg_alpha": 0.0, "reg_lambda": 1.0},
    {"max_depth": 6, "subsample": 0.8, "reg_alpha": 0.5, "reg_lambda": 1.0},
    {"max_depth": 6, "subsample": 0.8, "reg_alpha": 1.0, "reg_lambda": 2.0},
]

for params in param_settings:
    print("Training XGBoost with params:", params)

    model = XGBClassifier(
        n_estimators=best_n_estimators,
        learning_rate=best_lr,
        max_depth=params["max_depth"],
        subsample=params["subsample"],
        colsample_bytree=0.8,
        reg_alpha=params["reg_alpha"],
        reg_lambda=params["reg_lambda"],
        scale_pos_weight=xgb_scale_pos_weight,
        objective="binary:logistic",
        eval_metric="logloss",
        random_state=RANDOM_STATE,
        tree_method="hist",
        n_jobs=-1
    )

    start_time = time.time()
    model.fit(X_train_xgb, y_train, verbose=False)
    train_time = time.time() - start_time

    y_val_prob = model.predict_proba(X_val_xgb)[:, 1]
    best_threshold = get_best_threshold_by_f1(y_val, y_val_prob)
    y_val_pred = apply_threshold(y_val_prob, best_threshold)

    result = evaluate_model(
        name="XGB_tuned",
        y_true=y_val,
        y_pred=y_val_pred,
        y_prob=y_val_prob,
        threshold=best_threshold
    )
    result.update(params)
    result["train_time_sec"] = train_time
    xgb_tuning_results.append(result)

xgb_tuning_df = results_to_df(xgb_tuning_results)

best_xgb_row = xgb_tuning_df.iloc[0]
best_xgb_params = {
    "max_depth": int(best_xgb_row["max_depth"]),
    "subsample": float(best_xgb_row["subsample"]),
    "reg_alpha": float(best_xgb_row["reg_alpha"]),
    "reg_lambda": float(best_xgb_row["reg_lambda"]),
}
best_xgb_threshold = float(best_xgb_row["threshold"])

print("Best XGBoost params:", best_xgb_params)
print("Best XGBoost validation threshold:", best_xgb_threshold)

# =========================
# 13. Final XGBoost (retrain on train+val)
# =========================
print("\n" + "=" * 60)
print("Final XGBoost: retrain on train+val")
print("=" * 60)

preprocessor_xgb_final = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer_xgb, numeric_cols),
        ("cat", categorical_transformer_xgb, categorical_cols)
    ]
)

X_trainval_xgb = preprocessor_xgb_final.fit_transform(X_trainval)
X_test_xgb_final = preprocessor_xgb_final.transform(X_test)

feature_names_xgb_final = preprocessor_xgb_final.get_feature_names_out()

final_xgb = XGBClassifier(
    n_estimators=best_n_estimators,
    learning_rate=best_lr,
    max_depth=best_xgb_params["max_depth"],
    subsample=best_xgb_params["subsample"],
    colsample_bytree=0.8,
    reg_alpha=best_xgb_params["reg_alpha"],
    reg_lambda=best_xgb_params["reg_lambda"],
    scale_pos_weight=get_pos_weight(y_trainval),
    objective="binary:logistic",
    eval_metric="logloss",
    random_state=RANDOM_STATE,
    tree_method="hist",
    n_jobs=-1
)

start_time = time.time()
final_xgb.fit(X_trainval_xgb, y_trainval, verbose=False)
xgb_train_time = time.time() - start_time

xgb_test_prob = final_xgb.predict_proba(X_test_xgb_final)[:, 1]
xgb_test_pred = apply_threshold(xgb_test_prob, best_xgb_threshold)

xgb_test_result = evaluate_model(
    "Final_XGBoost",
    y_test,
    xgb_test_pred,
    xgb_test_prob,
    threshold=best_xgb_threshold
)
xgb_test_result["train_time_sec"] = xgb_train_time

print("Final XGBoost Test Result:")
display(pd.DataFrame([xgb_test_result]))

print("\nClassification report for Final XGBoost:")
print(classification_report(y_test, xgb_test_pred, zero_division=0))

# Feature importance table
xgb_importance = pd.DataFrame({
    "feature": feature_names_xgb_final,
    "importance": final_xgb.feature_importances_
}).sort_values("importance", ascending=False).reset_index(drop=True)

print("Top 20 important XGBoost features:")
display(xgb_importance.head(20))

plot_top_feature_importance(
    xgb_importance,
    top_n=20,
    title="XGBoost Feature Importance (Top 20)",
    filename="xgb_feature_importance_top20.png"
)

# =========================
# 14. MLP Experiment 1: architecture
# =========================
print("\n" + "=" * 60)
print("MLP Experiment 1: hidden_layer_sizes")
print("=" * 60)

mlp_arch_results = []
mlp_arch_loss_curves = {}

architectures = [
    (64,),
    (128, 64),
    (256, 128, 64)
]

for arch in architectures:
    print(f"Training MLP with hidden_layer_sizes={arch}")

    model = MLPClassifier(
        hidden_layer_sizes=arch,
        activation="relu",
        learning_rate_init=0.001,
        max_iter=100,
        random_state=RANDOM_STATE,
        early_stopping=False
    )

    start_time = time.time()
    model.fit(X_train_mlp, y_train)
    train_time = time.time() - start_time

    y_val_prob = model.predict_proba(X_val_mlp)[:, 1]
    best_threshold = get_best_threshold_by_f1(y_val, y_val_prob)
    y_val_pred = apply_threshold(y_val_prob, best_threshold)

    result = evaluate_model(
        name=f"MLP_arch_{arch}",
        y_true=y_val,
        y_pred=y_val_pred,
        y_prob=y_val_prob,
        threshold=best_threshold
    )
    result["architecture"] = str(arch)
    result["train_time_sec"] = train_time
    result["n_iter_"] = model.n_iter_
    mlp_arch_results.append(result)

    mlp_arch_loss_curves[str(arch)] = model.loss_curve_

mlp_arch_df = results_to_df(mlp_arch_results)

# plot using the exact same row order as architectures
arch_plot_df = pd.DataFrame({
    "architecture": [str(a) for a in architectures]
}).merge(
    mlp_arch_df[["architecture", "auc_pr"]],
    on="architecture",
    how="left"
)

plt.figure(figsize=(8, 5))
plt.plot(arch_plot_df["architecture"], arch_plot_df["auc_pr"], marker="o")
plt.title("MLP: Validation AUC-PR by Network Architecture")
plt.xlabel("hidden_layer_sizes")
plt.ylabel("Validation AUC-PR")
plt.grid(True)
save_current_figure("mlp_architecture_aucpr.png")
plt.show()

best_arch_row = mlp_arch_df.iloc[0]
best_arch_str = best_arch_row["architecture"]
architecture_map = {str(a): a for a in architectures}
best_arch = architecture_map[best_arch_str]
print("Best MLP architecture:", best_arch)

# =========================
# 15. MLP Experiment 2: activation / learning_rate_init / max_iter
# =========================
print("\n" + "=" * 60)
print("MLP Experiment 2: activation / learning_rate_init / max_iter")
print("=" * 60)

mlp_tuning_results = []

mlp_param_settings = [
    {"activation": "relu", "learning_rate_init": 0.001, "max_iter": 50},
    {"activation": "relu", "learning_rate_init": 0.01,  "max_iter": 50},
    {"activation": "relu", "learning_rate_init": 0.1,   "max_iter": 50},
    {"activation": "tanh", "learning_rate_init": 0.001, "max_iter": 50},
    {"activation": "tanh", "learning_rate_init": 0.01,  "max_iter": 50},
    {"activation": "relu", "learning_rate_init": 0.001, "max_iter": 100},
    {"activation": "relu", "learning_rate_init": 0.001, "max_iter": 150},
]

for params in mlp_param_settings:
    print("Training MLP with params:", params)

    model = MLPClassifier(
        hidden_layer_sizes=best_arch,
        activation=params["activation"],
        learning_rate_init=params["learning_rate_init"],
        max_iter=params["max_iter"],
        random_state=RANDOM_STATE,
        early_stopping=False
    )

    start_time = time.time()
    model.fit(X_train_mlp, y_train)
    train_time = time.time() - start_time

    y_val_prob = model.predict_proba(X_val_mlp)[:, 1]
    best_threshold = get_best_threshold_by_f1(y_val, y_val_prob)
    y_val_pred = apply_threshold(y_val_prob, best_threshold)

    result = evaluate_model(
        name="MLP_tuned",
        y_true=y_val,
        y_pred=y_val_pred,
        y_prob=y_val_prob,
        threshold=best_threshold
    )
    result["activation"] = params["activation"]
    result["learning_rate_init"] = params["learning_rate_init"]
    result["max_iter"] = params["max_iter"]
    result["train_time_sec"] = train_time
    result["n_iter_"] = model.n_iter_
    mlp_tuning_results.append(result)

mlp_tuning_df = results_to_df(mlp_tuning_results)

best_mlp_row = mlp_tuning_df.iloc[0]
best_mlp_params = {
    "activation": best_mlp_row["activation"],
    "learning_rate_init": float(best_mlp_row["learning_rate_init"]),
    "max_iter": int(best_mlp_row["max_iter"]),
}
best_mlp_threshold = float(best_mlp_row["threshold"])

print("Best MLP params:", best_mlp_params)
print("Best MLP validation threshold:", best_mlp_threshold)

# =========================
# 16. Final MLP (retrain on train+val)
# =========================
print("\n" + "=" * 60)
print("Final MLP: retrain on train+val")
print("=" * 60)

preprocessor_mlp_final = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer_mlp, numeric_cols),
        ("cat", categorical_transformer_mlp, categorical_cols)
    ]
)

X_trainval_mlp = preprocessor_mlp_final.fit_transform(X_trainval)
X_test_mlp_final = preprocessor_mlp_final.transform(X_test)

# make sure dense for MLP
if sparse.issparse(X_trainval_mlp):
    X_trainval_mlp = X_trainval_mlp.toarray()
if sparse.issparse(X_test_mlp_final):
    X_test_mlp_final = X_test_mlp_final.toarray()

final_mlp = MLPClassifier(
    hidden_layer_sizes=best_arch,
    activation=best_mlp_params["activation"],
    learning_rate_init=best_mlp_params["learning_rate_init"],
    max_iter=best_mlp_params["max_iter"],
    random_state=RANDOM_STATE,
    early_stopping=False
)

start_time = time.time()
final_mlp.fit(X_trainval_mlp, y_trainval)
mlp_train_time = time.time() - start_time

plt.figure(figsize=(8, 5))
plt.plot(final_mlp.loss_curve_)
plt.title("Final MLP: Training Loss Curve")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
save_current_figure("mlp_final_loss_curve.png")
plt.show()

mlp_test_prob = final_mlp.predict_proba(X_test_mlp_final)[:, 1]
mlp_test_pred = apply_threshold(mlp_test_prob, best_mlp_threshold)

mlp_test_result = evaluate_model(
    "Final_MLP",
    y_test,
    mlp_test_pred,
    mlp_test_prob,
    threshold=best_mlp_threshold
)
mlp_test_result["train_time_sec"] = mlp_train_time

print("Final MLP Test Result:")
display(pd.DataFrame([mlp_test_result]))

print("\nClassification report for Final MLP:")
print(classification_report(y_test, mlp_test_pred, zero_division=0))

# =========================
# 17. Final Comparison
# =========================
comparison_df = pd.DataFrame([xgb_test_result, mlp_test_result])[
    ["model", "accuracy", "precision", "recall", "f1", "auc_pr", "threshold", "train_time_sec"]
].sort_values("auc_pr", ascending=False).reset_index(drop=True)

print("\nGBDT vs MLP Comparison on Test Set")
display(comparison_df)

# =========================
# 18. Save Tables
# =========================
xgb_lr_df.to_csv(os.path.join(RESULTS_DIR, "xgb_learning_rate_results.csv"), index=False)
xgb_n_estimators_df.to_csv(os.path.join(RESULTS_DIR, "xgb_n_estimators_results.csv"), index=False)
xgb_tuning_df.to_csv(os.path.join(RESULTS_DIR, "xgb_tuning_results.csv"), index=False)
xgb_importance.to_csv(os.path.join(RESULTS_DIR, "xgb_feature_importance.csv"), index=False)
mlp_arch_df.to_csv(os.path.join(RESULTS_DIR, "mlp_architecture_results.csv"), index=False)
mlp_tuning_df.to_csv(os.path.join(RESULTS_DIR, "mlp_tuning_results.csv"), index=False)
comparison_df.to_csv(os.path.join(RESULTS_DIR, "final_model_comparison.csv"), index=False)

# =========================
# 19. Summary Prints
# =========================
print("\n" + "=" * 60)
print("SUMMARY")
print("=" * 60)
print(f"Best XGBoost learning_rate: {best_lr}")
print(f"Best XGBoost n_estimators: {best_n_estimators}")
print(f"Best XGBoost params: {best_xgb_params}")
print(f"Best XGBoost threshold: {best_xgb_threshold:.4f}")

print(f"\nBest MLP architecture: {best_arch}")
print(f"Best MLP params: {best_mlp_params}")
print(f"Best MLP threshold: {best_mlp_threshold:.4f}")

print("\nFinal Test Comparison:")
display(comparison_df)

print(f"\nAll result tables saved to: {RESULTS_DIR}")
print(f"All figures saved to: {FIGURES_DIR}")
```
